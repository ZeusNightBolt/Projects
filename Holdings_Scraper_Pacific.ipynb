```python
list_of_DF = []
for i in range(0, len(dfs)):

  DF = pd.DataFrame(dfs[i])
  DF = DF.iloc[3:]
  DF = DF.drop(columns = [ 'Unnamed: 0', '6'])
  DF = DF.dropna(axis=1, how='all', thresh=None, subset=None, inplace=False)  
 

  list_of_DF.append(DF)

Combined_DataFrame = pd.concat(
    list_of_DF,
    axis=0,
    join="outer",
    ignore_index=True,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)

Combined_DataFrame.columns = ['Sec ID', 'Foreign Code', 'Date Acquired', 'Vendor', 'Actual Cost', 'Par Value', 'Paid for Accrued Interests and Dividends']


```

```python
excelfilename = r'Pacific_Holdings.xlsx'
Combined_DataFrame.to_excel(excelfilename, sheet_name = 'Aggregated_Offers', index=False)
```

```python

Combined_DataFrame['Actual Cost'] = Combined_DataFrame['Actual Cost'].astype(str)
Combined_DataFrame['Actual Cost'] = Combined_DataFrame['Actual Cost'].apply(lambda x: x.replace(',', ''))
Combined_DataFrame['Actual Cost'] = Combined_DataFrame['Actual Cost'].astype(dtype = float, errors = 'ignore')

Combined_DataFrame
```

```python
Combined_DataFrame_Temp = Combined_DataFrame.copy()

Combined_DataFrame_Temp = Combined_DataFrame_Temp[(Combined_DataFrame_Temp.Vendor != 'EXCHANGE')]
```

```python
Aggregated_Pivot = pd.pivot_table(Combined_DataFrame, values=['Actual Cost'], index=\
                              ['Sec ID',  'Date Acquired', 'Vendor'])
```

```python

Aggregated_Pivot = Aggregated_Pivot.reindex(Aggregated_Pivot['Actual Cost'].sort_values(ascending=False).index)
```

```python
import openpyxl
from openpyxl import load_workbook
import xlsxwriter
# Pivot_table1.to_excel(excelfilename, sheet_name="Aggregrated_Offers", startrow=7000)
book = load_workbook(excelfilename)
writer = pd.ExcelWriter(excelfilename, engine='openpyxl') 
writer.book = book
## ExcelWriter for some reason uses writer.sheets to access the sheet.
## If you leave it empty it will not know that sheet Main is already there
## and will create a new sheet.
P1 = book.create_sheet("CUSIP_SortedActualCost")
writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
Aggregated_Pivot.to_excel(writer, "CUSIP_SortedActualCost", startrow=1)
P1.column_dimensions['A'].width = 18
P1.column_dimensions['B'].width = 25
P1.column_dimensions['C'].width = 35
P1.column_dimensions['D'].width = 18
writer.save()
```
